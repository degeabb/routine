```python
def rename_col_list(startyear,endyear):
    l1=[]
    a = int(startyear)
    for i in range(int(endyear)-int(startyear)+1):
        l1.append(str(a)+ ' No of Funds')
        l1.append(str(a)+' Amount(US$m.)')
        a = a+1
    return(l1)
```

```python
col_list = rename_col_list(2013,2016)
col_list 
```

```python
df4.columns = col_list
```

```python
df4.head()
```

```python
def replace_col_list(startyear,endyear):
    l1=[]
    a = int(startyear)
    b = int(startyear)
    for i in range(int(endyear)-int(startyear)+1):
        l1.append(str(a)+ ' No of Funds')
        a = a+1
    for i in range(int(endyear)-int(startyear)+1):
        l1.append(str(b)+' Amount(US$m.)')
        b= b+1
    return(l1)
```

```python
replace_list = replace_col_list(2013,2016)
replace_list
```

```python
df4 = df4[replace_list]
```
