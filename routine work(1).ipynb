```python
import pandas as pd
import numpy as np
import os,sys
```

```python
df = pd.read_excel('D:/allpeexit.xlsx')
```

```python
df.head()
```

```python
print (df.columns)
```

```python
df['year'] = pd.DatetimeIndex(df['Transaction Date']).year
```

```python
def lam1(x):
    return x.isna().sum() + x.notna().sum()
def lam2(x):
    return x.notna().sum()
```

```python
table = pd.pivot_table(df, values=['Transaction Amount (US$m.)'], 
                                   index=['year', 'Investee/Target - Country'], 
                                   columns=['Exit Type'], 
                                   aggfunc=  [lam1, lam2, np.sum])
```

```python
table.columns.set_levels(['No. of Deals','No. of Disclosed Deals', 'US$'],level=0,inplace=True)
```

```python
table.columns = table.columns.droplevel(1)
table
```

```python
table.columns = table.columns.swaplevel(0, 1)
```

```python
table.sort_index(axis='columns',level = 'Exit Type', inplace = True)
```
